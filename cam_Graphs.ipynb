```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
from pathlib import Path
```

```python
readpath = Path.cwd() / 'Resources' / 'mergedData.csv'
df = pd.read_csv(readpath)
```

```python
np.unique(df['artist'])
```

```python
df.drop(['Unnamed: 0', 'Unnamed: 0.1', 'Region', 'Year', 'trackID', 'name', 'artist'], axis=1, inplace=True)
```

```python
df.columns
```

```python
streamCount = df.groupby('Artist')[['Artist', 'TrackID', 'Streams', 'danceability', 'acousticness', 'energy', 'liveness', 'loudness', 'speechiness', 'tempo']].sum()
```

```python
streamCount_df = pd.DataFrame(data=streamCount)
streamCount_df.head()
```

```python
maxArtist = streamCount_df.idxmax()[0]
minArtist = streamCount_df.idxmin()[0]
print(f'The artist with the max streams is {maxArtist}. The artist with the minimum streams is {minArtist}.')
```
