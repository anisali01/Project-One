```python
# Import 
from pathlib import Path
import pandas as pd
```

```python
# Load files
readpath = Path.cwd() / 'Resources' / 'data.csv'
readpath
```

```python
spotipyDF = pd.read_csv(readpath, encoding = "ANSI")
```

```python
spotipyDF.head()
```

```python
spotipyDF = spotipyDF.dropna(how='any')
```

```python
# Clean any misc. rows that contain questionable artist names

spotipyDF = spotipyDF[spotipyDF.Artist != "Z"]
spotipyDF = spotipyDF[spotipyDF.Artist != "13 Block"]
spotipyDF = spotipyDF[spotipyDF.Artist != "10cm"]
spotipyDF = spotipyDF[spotipyDF.Artist != "070 Shake"]
spotipyDF = spotipyDF[spotipyDF.Artist != "6150"]
spotipyDF = spotipyDF[~spotipyDF['Artist'].str.contains('\?')]
spotipyDF = spotipyDF[spotipyDF.Artist != "?"]
spotipyDF = spotipyDF[spotipyDF.Artist != "100%"]
spotipyDF
```

```python
# Clean any misc. rows that contain questionable track names

spotipyDF = spotipyDF[spotipyDF.TrackName != ""]
spotipyDF = spotipyDF[~spotipyDF['TrackName'].str.contains('\?')]
spotipyDF
```

```python
# Filter data to only use US data in 2017

spotipyDF = spotipyDF[spotipyDF["Region"] == "us"]
spotipyDF
```

```python
spotipyDF['Year'] = spotipyDF['Date'].str.extract('.*(\d{2})', expand = False)
spotipyDF.head()
```

```python
# Create buckets of streaming values

bins = [0, 50000, 250000, 500000 , 750000, 1000000, 1500000, 2000000, 2500000, 9999999]
binNames = ["<50,000","50,001 - 250,000", "250,001 - 500,000", "500,001 - 750,000", "750,001 - 1,000,000", "1,000,001 - 1,500,000", "1,500,001 - 2,000,000", "2,000,001 - 2,500,000", ">2,500,000"]
spotipyDF["bins"] = pd.cut(spotipyDF["Streams"], bins, labels = binNames)
grouping = spotipyDF.groupby("bins")
spotipyDF
```

```python
spotipyDF = spotipyDF[spotipyDF.Year != '18']
```

```python
printpath = Path.cwd() / 'Resources' / 'cleanData.csv'
spotipyDF.to_csv(printpath)
```
